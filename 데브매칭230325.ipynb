```python
wall = ["3 1 1 2", "2 3 2", "1 2 3 1"]
def solution(wall):
    wall = [list(map(int, x.split())) for x in wall]
    graph = [[None] * sum(wall[0]) for _ in range(len(wall))]
    for row, row_data in enumerate(wall):

        cnt = 0
        for col, col_data in enumerate(row_data):
            for idx in range(col_data):
                graph[row][cnt + idx] = (row, col)
            cnt += col_data
    print(graph)
    input()
    nums = {}
    
    dx = [1, 0, -1, 0]
    dy = [0, 1, 0, -1]

    for row in range(len(graph)):
        for col in range(len(graph[row])):
            if graph[row][col] not in nums:
                nums[graph[row][col]] = set()
            for x, y in zip(dx, dy):
                x += row
                y += col
                if x < 0 or y < 0:
                    continue
                if x > len(graph) - 1 or y > len(graph[row]) - 1:
                    continue
                
                print(x, y, len(graph[row]) - 1, row, col)
                print(graph[row][col])
                print(graph[x][y])
                if graph[row][col] != graph[x][y]:
                    nums[graph[row][col]].add(graph[x][y])

    max_len = max(list(map(len, list(nums.values()))))

    answer = []
    for key, value in nums.items():
        if len(value) == max_len:
            answer.append(list(key))
    return answer

solution(wall)
```
